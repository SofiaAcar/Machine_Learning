```python
R2_score = corr ** 2
R2_score
```

## Linear Regression

```python
sns.regplot(x = 'total_spend', y = 'sales', data = df, ci = None);
```

Best fit line' ı elde etmiş olduk. Best line' ın üstünde ve altında kalan değerlerin toplamı her zaman 0' ı verir. Bu yüzden gerçek değer ile tahmin edilen değer arasındaki farkların karesi alınır. (Bazı kaynaklarda karesi yeerine mutlak değer alınır.)

## Splitting the dataset into X(independent variables) and y (dependent variable)

```python
# y_pred = b1X + b0
```

```python
X = df.total_spend      # Independent variable
y = df.sales            # Dependent variable
```

## Determination of coefficients (slope and intercept)

```python
np.polyfit(X, y, deg=1)
```

Bulunan ilk değer b1 (slope), ikinci değer b0 (intercept)' dır.

```python
slope, intercept = np.polyfit(X, y, deg=1)
```

```python
print('slope :', slope)
print('intercept :', intercept)
```

## Why do we use the least squares error method to find the regression line that best fits the data?

```python
b1, b0 = np.polyfit(X, y, deg=1)
```

```python
print('b1 :', b1)
print('b0 :', b0)
```

```python
y_pred = b0 + b1 * X
```

```python
values = {'actual' : y, 'predicted' : y_pred, 'residual' : (y - y_pred), 'LSE' : (y - y_pred) ** 2}
df_2 = pd.DataFrame(values)
df_2
```

Least Squared Errors (LSE)' da kareleri almazsak değer 0 çıkıyor. Bu yüzden kareler üzerinden işlem yapıyoruz:

```python
df_2.residual.sum().round()
```

```python
df_2.LSE.sum()   # Çizilen birçok LSE' ler arasındaki en küçük score olduğunu biliyoruz. 
                 # 'Best line'
```

## Prediction with simple linear regression

```python
potential_spend = np.linspace(0, 500, 100)
potential_spend
```

Modelin bir tahmin yapması için potential_spend'i X olarak tanımladık.

```python
predicted_sales_lin = b0 + b1 * potential_spend
predicted_sales_lin
```

```python
plt.plot(potential_spend, predicted_sales_lin)
sns.scatterplot(x = 'total_spend', y = 'sales', data = df);
```
