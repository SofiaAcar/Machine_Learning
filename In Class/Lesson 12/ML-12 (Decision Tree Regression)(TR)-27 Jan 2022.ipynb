## Modeling

DesicionTreeRegression mantigi : 'Ilk nerden bolersem iki taraftaki variance' lar en kucuk olur?' mantigi ile ayrim yapar. Yani bolgeler arasi variance' i en aza indirmeye calisir. Sayilar arasi farki en az olanlar ayni sinifa duserler. 

!!! VARIANCE uzerinden sinif ayrimi yapar; MEAN uzerinden predict yapar. !!!

Variance kucuk olsun istiyoruz ama hataya hic izin verilmezse model datayi ezberler. Bu da bizi overfitting' e goturur.

!!!! DesicionTreeClassification, Gini ve Entropy metrikleri ile calisiyordu; DesicionTreeRegression, MSE ile calisir.!!!!

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```python
DT_model = DecisionTreeRegressor(random_state=101)
```

```python
DT_model.fit(X_train, y_train)
```

```python
y_pred = DT_model.predict(X_test)
y_train_pred = DT_model.predict(X_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
def train_val(model, X_train, y_train, X_test, y_test):
    
    y_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    
    scores = {"train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    "test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
train_val(DT_model, X_train, y_train, X_test, y_test)
```

Yukaridaki skorlarda bir overfitting durumu oldugunu goruyoruz. Model, train setindeki tahminleri %100 dogru bilmis.

__Cross_Validate__

```python
from sklearn.model_selection import cross_validate, cross_val_score
model = DecisionTreeRegressor(random_state=101)
scores = cross_validate(model, X_train, y_train, scoring=['r2', 
            'neg_mean_absolute_error','neg_mean_squared_error','neg_root_mean_squared_error'], cv =5)
df_scores = pd.DataFrame(scores)
df_scores.mean()[2:]
```

Cross validate isleminden sonra skorlarin kotulestigini gorduk. Demek ki burada gercekten bir overfitting durumu var.

```python
def report_model(model):
    y_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    print('\n')
    print(train_val(model, X_train, y_train, X_test, y_test))
    print('\n')
    plt.figure(figsize=(12,8),dpi=100)
    plot_tree(model,filled=True, feature_names=X.columns, fontsize=10);
```

```python
pruned_tree = DecisionTreeRegressor(max_depth=3, random_state=101)
pruned_tree.fit(X_train,y_train)
```

Classification' da bolme metrigi : Gini ve Entropi

Regression' da bolme metrigi : Mean Squared Error (Variance)

Amacimiz; mean squared error' u 0' a yaklastirmak. 

Classification' da ayni class' ta olanlara gore bir ayrim yapiyordu; simdi fiyatlari birbirine yakin olanlara gore bir ayrim yapacak.

Ilk daldaki value=4.574 : Tum targettaki sample larin ortalama fiyatidir.

Ikinci dalda fiyati cok yuksek olanlari sag dala (value=23.499), dusuk olanlari sol dala (value=3.922) ayirmis.

```python
report_model(pruned_tree)
```

En sagdaki leaf' lerden ornek bir hesaplama yapalim : value=34.0 olan dal value=33.0 ve value=35.0 olarak ayrilmis. burdaki hesap su sekilde yapilir :

[(33-35) ** 2 + (34-35) ** 2] / 2 = 1   (Gozlem sayisina bolduk.) Boylece mse' yi 1 bulmus olduk.

Yukarida max_depth dedigimiz icin 3 adim gitti ve durdu. En son dalda kalan arabalar ortalama olarak ayni segmentteler fiyatlari birbirine yakin. Bu yuzden daha fazla ilerlemeye gerek yok. Boylece overfitting' i engellemis olduk ve model bir genelleme yapabildi.

## GridSearch

```python
from sklearn.model_selection import GridSearchCV
```
